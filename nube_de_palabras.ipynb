* El siguiente análisis es una herramienta de Minería de Texto denominada Nube de Palabras, que tiene como objetivo visualizar las palabras del texto en un gráfico donde las palabras que mas se repiten se ordenan de mayor tamaño. El texto analizado es el Apolcalipsis de la Biblia.

```python
# Importamos las librerías necesarias

import spacy
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Cargamos el archivo de texto

with open("APOCALIPSIS.txt", 'r', encoding="latin1") as file:
    text = file.read()

wordcloud = WordCloud().generate(text)
```

```python
# Generamos la imagen

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# Generamos una segunda imagen, esta vez limitando el tamaño máximo de palabra

wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Cargamos las stop words en español 
es_stopwords = open("stopwords-es.txt").read().splitlines()

stopwords = set()
stopwords.update(es_stopwords)

# Generate una images de word cloud, esta vez fijamos el color de fondo en blanco
wordcloud = WordCloud(stopwords=stopwords, max_font_size=60, background_color="white").generate(text)

# Desplegamos la imagen usando matplotlib
plt.figure(dpi=150)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
